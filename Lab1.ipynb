```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns   
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler
```

```python
data=[10,20,30,40,50]
mean= np.mean(data)
median= np.median(data)
mode= pd.Series(data).mode()[0]
```

```python
print(mean)
print(median)
print(mode)
```

```python
np.random.seed(42)
normal_data = np.random.normal(loc=50, scale=10, size=1000)
```

```python
plt.figure(figsize=(9,5))
sns.histplot(normal_data, kde=True, bins=30, color='blue')
plt.title("Distribution of Normal Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
df=pd.DataFrame({
    "Fruit": ["Apple", "Banana", "Orange", "Apple", "Banana", "Apple"]
})

label_encoder=LabelEncoder()
df["Fruit_LabelEncoded"]= label_encoder.fit_transform(df["Fruit"])

one_hot=pd.get_dummies(df["Fruit"], prefix="Fruit")

print(df[["Fruit"]])
print("\n Label Encoded")
print(df[["Fruit", "Fruit_LabelEncoded"]])
print("\n One-Hot Encoded")
print(one_hot)
```

```python
df2=pd.DataFrame({
    "Height_cm": [150, 160, 170, 180, 190],
    "weight_kg": [50, 60, 70, 80, 90]
})

print("Original Data:\n",df2)

minmax_scalar=MinMaxScaler()
df2_minmax= pd.DataFrame(minmax_scalar.fit_transform(df2), columns=df2.columns)

standard_scalar=StandardScaler()
df2_standard=pd.DataFrame(standard_scalar.fit_transform(df2), columns=df2.columns)

print("\n Min-Max Scaled Data:\n", df2_minmax)
print("\n Standard Scaled Data:\n", df2_standard)
```
